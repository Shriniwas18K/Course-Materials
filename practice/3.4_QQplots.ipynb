###### 2] QQ-plots ( widely used )

###### QQ-plot method

##### say we have distribution X and we want to determine its type of distribution wheter it is normal or not, so we will do this process as below
##### step1 : generate a normal distribution Y using random module inside numpy
##### step2 : sort X and Y ,and calculate 1 to 100 percentiles of both them seprately
##### step3 : plot a scatterplot between these percentile values 
##### step4 : if maximum number of points lie very near or on a best fit line , then X also follows normal distribution

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
iris=load_iris()
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()
```

```python
Y=np.random.normal(loc=5,scale=1,size=150)
X=df['sepal length (cm)']
```

```python
Ypercentiles=[]
Xpercentiles=[]
for i in range(1,101):
    Xpercentiles.append(np.percentile(X,i))
    Ypercentiles.append(np.percentile(Y,i))
```

```python
sns.scatterplot(y=Ypercentiles,x=Xpercentiles)
```

##### above plot shows that maximum points can be fit to a best fit line , so its sort of normal distribution but not completely normal distribution
##### we dont need to do this manually instead we cant just focus on identifying type of distribution , use statsmodel library with advanced functionalities of fitting best fit line there

###### line='45' is always used

```python
import statsmodels.api as sm
fig=sm.qqplot(df['sepal length (cm)'],line='45',fit=True)
plt.ylabel('sepal length (cm) values')
plt.xlabel('theoritical normal values')
plt.title('QQplot of sepal length (cm)')
```
