```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("StudentsPerformance.csv")
df.head(3)
```

##### group by is similar to those group by in databases , it segragates rows of data frame on basis of particular categorical columns

```python
df.info()
```

##### here gender column is our basis of grouping

```python
grps=df.groupby('gender')
```

##### now internally the data is segragated into diffirent dataframes on basis of values of gender , now we can apply various aggregate and mathematical functions on these DataFrames like mean ,median , std , var ,describe, etc , Basically these all are numeric functions , and hence they work only on numeric data ,  so we can  also say that we are trying to segragate our numeric data into various DataFrames on the basis of values in the Categorical features using groupby
##### we can obtain all numeric features only grouped dataframes if we group by all categorical columns else , our grouped dataframes still contain other categorical columns which were not basis of grouping

```python
grps.mean(numeric_only=True)
```

```python
grps.describe()
```

```python
df.mean(numeric_only=True)
```

```python
df.describe()
```

##### see the diffirence of using groupby , we are getting our results grouped on basis of gender value

##### we can use group by with multiple categorical features , or ordinal features

```python
grps1=df.groupby(['gender','race/ethnicity','test preparation course'])
```

```python
grps1.mean(numeric_only=True)
```

```python
grps1.describe()
```

### see here we are getting dataframe as a result , but it has two or more index columns on basis of how many columns we have grouped , such DataFrame is called as MultiIndex DataFrame

##### we can access values inside the Dataframes of the groups by the way of accessing Multiindex DataFrame
