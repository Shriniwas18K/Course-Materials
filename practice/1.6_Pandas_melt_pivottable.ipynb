```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("details.csv")
df.head(3)
```

##### generally there are two methods to store data , 
##### 1] long data : no. of rows is extremely high than number of columns (such data we recieve from databases ) and
##### 2] wide data : no. of columns is extremely high than number of rows ( such data we recieve from datalakes , datawarehouses,webapis,webscraping)

```python
df.shape
```

##### this shape indicates no.of rows=17612 , no. of columns=45 , this dataset is long data form 

#### melt method converts any dataframe to its maximum extent to become long data form

```python
df.melt()
```

##### pivot_table method converts any dataset to maximum extent of wide data form , but moreover it is advanced version of groupby with more functionality , and acts as shorter method instead of writing multiple groupby statements , hence used widely

```python
df.pivot_table(columns=['ball'],aggfunc='mean')
```

###### this above pivot table shows avg of every feature for each ball , i.e. same could be done as below

```python
grps=df.groupby(['ball'])
grps.mean().T.sort_index()
```
