```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('xAPI-Edu-Data.csv')
df.head()
```

##### segragating boxplot of numerical feature on basis of values in categorical feature

```python
df.plot(kind='box',by='StageID',column='raisedhands')
```

##### pointplot of numerical variable segragted using categorical feature values

```python
sns.catplot(data=df,x='StageID',y='raisedhands',kind='point')
```

##### bar chart of categorical features values based on its numeric feature values

```python
sns.catplot(data=df,x='StageID',y='raisedhands',kind='bar')
```

##### boxenplot of numerical variable based on categorical feature values

```python
sns.catplot(data=df,x='StageID',y='raisedhands',kind='boxen')
```

##### voilinplot of numerical variable based on categorical feature values

```python
sns.catplot(data=df,x='StageID',y='raisedhands',kind='violin')
```
