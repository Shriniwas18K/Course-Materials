```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### random variable : it is set . i.e. collection of outcomes of a random experiment , denoted by Capital letters
###### random experiment : an event whose outcome we know but we cant actually predict exactly what will it be , for example : rolling of dice : we know its outcomes can be only 1,2,3,4,5,6 but we cant predict exactly what will come , hence here  random variable X = { 1,4,6,2,2,5,3,4,5,6  } for 10 rolls and all the unique elements in it are togeherly known as sample space(all possible outcomes of an event)
##### there are two types of random variables : discrete and continous :    
###### discrete random variable : its sample space contains all distinct values whose probability can be found and not equal to 0 . ex: rolling of dice : each outcome gas definite probability of 1/6 
###### continous random variable : its sample space contains infinite number of values but between fixed range , i.e. infinite number of outcomes , and if we try to calculate probability of each individual outcome then it is 0 , ex. jee percentile value of students : it can be anything between 0 to 100 , say 98.3534463463453, but probability of exactly getting this percentile from all students is practically 0 , i.e. such cases where outcomes are floating point numbers ex. X={ 94.345345 , 92.45345345 , 90.3435 } etc , or in other case it can contain any integer value between very large range so as to the probability of occurence of any individual value is almost 0 , ex. X={10,900,54,35,90,234,51433,4580,45023,45,2,5} i.e. its sample space is infinite number of integers 

##### straight forwardly : discrete random variable has finite sample space , and continous random variable has infinite sample space
### In this world we some commonly appearing probability distributions like normal,log normal , binomial , bernoulli , etc ,so whenever we get any random variable , then we plot its histogram, if that looks similar to any of those commonly found distributions then we consider that the random variable follows all properties of that distribution hence we get a lot of information using that distribution properties applicable on our random variable
##### solved ambiguity : histograms are frequency vs outcome plot , and probability distributions are probability vs outcome plots , then how do establish the relation that random variable follows the same properties of them , ans is that probability of an outcome is actually frequency of outcome divided by cummulative frequency where cummulative frequency indirectly means , total no. of times event occured or sayingly total no. of outcomes(including duplicate outcomes)

### NOTE : here binning cannot be used at all , due to further complex applications of PMF and PDF

### here we are just illustrating for Xdis , but this can be for any event with discrete random variable 
##### first case of below random variable Xdis contains outcomes of rolling of dice

```python
Xdis=np.random.randint(1,7,256)
Xdis
```

```python
sns.histplot(Xdis,color='m')
```

##### for any discrete random variable we can plot its histogram , and bar chart perfectly

```python
pd.Series(Xdis).value_counts().sort_index()
```

##### above is frequency distribution table for 256 rolls of dice

```python
k=pd.Series(Xdis).value_counts().sort_index()/256
k
```

##### above is probability distribution table for it

```python
k.plot(kind='bar',color='m')
```

### the probability vs outcome plot for discrete random variable is called probability mass function(PMF)
#### see that is similar with its histogram

##### theoritically probability of each outcome of rolling of dice is equal to 1/6 but practically it may vary a little above as it is from outcomes in random experiment , but still sum of probabilities of all outcomes remains 1

```python
k
```

```python
k.sum()
```

##### we can plot cummulative distribution function (CDF) , which is cummulative sum of all probabilites for each and its lower outcome values

```python
k.cumsum()
```

```python
sns.barplot(x=k.index,y=k.cumsum())
```

### lets see case of continous random variable : 

```python
Xcon=np.random.rand(100)*100
Xcon
```

##### for such data , we cant have frequency distribution table useful as each value has occured only once ot twice , so our last option for frequency distribution is binning and then we can have histogram , we can adjust bin width and bins parameter

```python
sns.histplot(Xcon,bins=15,color='m')
```

##### for such continous random variable we need to plot probability density function ( PDF ) , which can be plotted by looking at histogram
### Case 1: if our histogram matches normal distribution(commonly found famous distribution known as "normal distribution " ) , then we plot the PDF using the formula given by 
##### P=1/(2(rootpi))* e^( (x -mean)/std ) 
##### where P=probability density value of data point, x=data point

### Case 2: if our histogram does not match any of the commonly appearing probability distributions , then we have to use kernal density method to plot its PDF
##### kernal density estimation steps: we will use gaussian kernal
##### step 1 : we draw gaussian distribution around each datapoint considering it as mean , the standard deviation of this curve is called as bandwidth of kernal , by decreasing its value the curve becomes more spikey tall , increasing becomes more small and wide
##### step 2: the probability density value of the datapoint is the sum of all probability densities which intersect when a vertical line is drawn through the data point
##### step 3: follow this with each datapoint and we get the PDF of the data
##### this all process is built in inside seaborn library histplot and kdeplot functions , and there we can also modify bandwidth value hence no need of doing everything manually , we have to adjust bandwidth value using methods bw_method , bw_adjust , any one of them can be used  

#### above histogram does not match any of commonly found probability distributions so we will consider it Case 2 and plot its PDF using kernal density estimation method

```python
sns.histplot(Xcon,kde=True,bins=15,color='m')
```

##### see now decreasing bandwidth value that kde curve will make itself more spikey , bin width can be adjusted using bin_width parameter , and number of bins can be adjusted by bins parameter

```python
sns.histplot(Xcon,kde=True,color='m',kde_kws={"bw_method":0.1},bins=15)
```

##### only its PDF curve without bars can be plotted using kdeplot method

```python
sns.kdeplot(Xcon,bw_method=0.1,fill='g')
```

### PDF curve of Xcon continous random variable , if it matches any of commonly found distributions then we can directly infer that distribution's properties to this random variable

##### now we can calculate probability using area under above curve by integrating between desired range 

##### we cannot calculate the probability for individual data points for continous random variable as it is actually 0 , by probability that data points lie within given range is non-zero and finite calculatable

##### we can plot cummulative density function ( CDF ) of continous random variable as below : it contains cummulative density on y-axis

```python
sns.kdeplot(Xcon,cumulative=True,fill='g')
```

##### here we can calculative cummulative probability between given range by integrating in above curve

#### descriptive stats measures are not at all affected by plotting PDF PMF of random variable ,  i.e. as we are adjusting bandwidth etc so shape of curve changes little, but variance stddev skewness kurtosis of variable are not at all affected
