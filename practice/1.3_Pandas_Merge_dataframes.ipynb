### Most of the times in real world , we need to work by  merging various DataFrames , as Data comes from various sources like Database, APIs , DataLakes , DataWarehouses , etc

##### so to work with those all Data , we need to Merge them all into single DataFrame

### here we involve concepts of SQL JOINS : inner , left outer , right outer , outer , full outer , cross ,etc joins
##### merge function is used to perform joining of two tables at one time

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("StudentsPerformance.csv")
df=df[:10:]
df.head(3)
```

```python
df.shape
```

#### we have made an imaginary dataframe for our use of merging demonstration : we are using pd.concat() funcn to make dataframe writing axis=1, else axis=0 will concat just series to series and not make dataframe ,
#### similarly we can vertically add new rows to our dataframe using df.append() funcn

```python
arr1=np.random.randint(10,100,8)
arr2=np.random.randint(50,100,7)
```

```python
import random
d=['data analyst','data scientist','startup','mlops','mlengineer']
l=[]
for a in range(6):
    l.append(np.random.choice(d))
arr3=np.array(l)
```

```python
s1=pd.Series(arr1)
s2=pd.Series(arr2)
s3=pd.Series(arr3)
```

```python
ndf=pd.concat((s1,s2,s3,df['math score'].iloc[:5:]),axis=1)
ndf.rename(columns={0:'science score',1:'eng score',2:'role',3:'math score'},inplace=True)
ndf.head(3)
```

```python
ndf.shape
```

#### now our this new DataFrame contains 900 rows , where some columns contain again missing values 
#### and our original DataFrame df contains 1000 rows all filled
## both DataFrames have diffirent columns,but atleast one column should be common exactly same between them to perform operation of merge and joins in it

#### cross join using merge 

```python
pd.merge(left=df,right=ndf,how='cross')
```

##### inner join : intersection of two tables using merge : we are displaying those observations where math score from both tables is same

```python
pd.merge(left=df,right=ndf,how='inner',on='math score')
```

##### right outer join : we are displaying whole ndf table at right and only those observations in left df table where math score is same : else where math score in left table does not match but still extra rows exist in right table , then we displaying those corresponding space of left side as NaN values

```python
pd.merge(left=df,right=ndf,how='right',on='math score')
```

##### left outer join : whole reverse case of right join : we will display whole left df table , and displaying only those observations of right table ndf where math score is same, else NaN values are displayed in right side

```python
pd.merge(left=df,right=ndf,how='left',on='math score')
```

##### union of two tables can be displayed as , where number of observations diffiers in that side NaN values are filled

```python
pd.merge(left=df,right=ndf,how='outer')
```

##### this we have done on basis of only one column math score , but we can use more than one columns by passing list of columns in "on" parameter
